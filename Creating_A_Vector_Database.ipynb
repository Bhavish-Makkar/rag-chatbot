**Importing The Required Libraries**

```python
import json
import tiktoken
import gdown

import pandas as pd

from langchain.text_splitter import RecursiveCharacterTextSplitter

from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain_community.embeddings.sentence_transformer import (
    SentenceTransformerEmbeddings
)
from langchain_community.vectorstores import Chroma
```

# **Creating My Vector Database**

**Document Loading**

```python
pdf_folder='/content/RAG_Project'
```

```python
pdf_loader = PyPDFDirectoryLoader(pdf_folder)
```

**Chunking of the Document**

```python
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    encoding_name='cl100k_base',
    chunk_size=512,
    chunk_overlap=16
)
```

```python
docs_chunks = pdf_loader.load_and_split(text_splitter)
```

**Creation of Vector Database**

```python
embedding_model = SentenceTransformerEmbeddings(model_name='Qwen/Qwen3-Embedding-0.6B')
```

```python
document_collections="Assignment_documents"
```

```python
pip install --upgrade transformers sentence-transformers accelerate
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("Qwen/Qwen3-Embedding-0.6B")
```

```python
vectorstore = Chroma.from_documents(
    docs_chunks,
    embedding_model,
    collection_name=document_collections,
    persist_directory='./docs_db'
)
```
